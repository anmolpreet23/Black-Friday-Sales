# BLACK FRIDAY SALE ANALYSIS

Black Friday marks the beginning of the Christmas shopping festival across the US.This project basically aims at understanding the customer purchase behaviour.Here we have the purchase summary of various customers . The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

```python
import tkinter as tk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
data=pd.read_csv('train.csv')
data['Purchase'].fillna(0,inplace=True)
data['Purchase'].replace(0,"None",inplace=True)
data.update(data[['Age','Occupation','Marital_Status']].fillna(0))


def world():
    sexRatio=data.groupby('Gender').size()
    root=tk.Tk() 
    labels = 'F', 'M'
    sizes = sexRatio
    explode = (0, 0.1)
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    bar1 = FigureCanvasTkAgg(fig1, root)
    bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    figure1 = plt.Figure(figsize=(16,16), dpi=100)
    ax1.set_title('Male vs Female Ratio')
    
    root.mainloop()
```

```python
def mworld():
    mRatio=data.loc[(data['Gender']=='M')].groupby(data['Stay_In_Current_City_Years']).size()
    root= tk.Tk() 
    mRatio.head()
    figure1 = plt.Figure(figsize=(16,12), dpi=100)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, root)
    bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    mRatio.plot(kind='bar', ax=ax1)
    ax1.set_title('Years males have stayed in the city')
    root.mainloop()
```

```python
def fworld():
    g=data.loc[(data['Marital_Status']==0)].groupby('Occupation').size()
    root= tk.Tk() 

    figure1 = plt.Figure(figsize=(16,8), dpi=100)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, root)
    bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    g.plot(kind='bar', ax=ax1)
    ax1.set_title('Unmarried vs Occupation')
    root.mainloop()
```
